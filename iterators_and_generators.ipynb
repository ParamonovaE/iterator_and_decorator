```python
class FlatIterator:
    def __init__(self, list_of_lists):
        self.list_of_lists = list_of_lists
        self.main_index = 0
        self.sub_index = 0

    def __iter__(self):
        return self

    def __next__(self):
        #Если не конец списка
        if self.main_index < len(self.list_of_lists):
            # Получаем текущий подсписок
            current_list = self.list_of_lists[self.main_index]
            # Если подсписок закончился, то переходим к следующему
            if self.sub_index >= len(current_list):
                self.main_index += 1
                self.sub_index = 0
                return self.__next__()
            # Если подсписок не закончился, то возвращаем элементы из него
            else:
                item = current_list[self.sub_index]
                self.sub_index += 1
                return item    
        else:
            raise StopIteration

def test_1():

    list_of_lists_1 = [
        ['a', 'b', 'c'],
        ['d', 'e', 'f', 'h', False],
        [1, 2, None]
    ]

    for flat_iterator_item, check_item in zip(
            FlatIterator(list_of_lists_1),
            ['a', 'b', 'c', 'd', 'e', 'f', 'h', False, 1, 2, None]
    ):

        assert flat_iterator_item == check_item

    assert list(FlatIterator(list_of_lists_1)) == ['a', 'b', 'c', 'd', 'e', 'f', 'h', False, 1, 2, None]


if __name__ == '__main__':
    test_1()
```

```python
import types
def flat_generator(main_list):
    for sub_list in main_list:
        for item in sub_list:
                yield item
        
def test_2():

    list_of_lists_1 = [
        ['a', 'b', 'c'],
        ['d', 'e', 'f', 'h', False],
        [1, 2, None]
    ]

    for flat_iterator_item, check_item in zip(
            flat_generator(list_of_lists_1),
            ['a', 'b', 'c', 'd', 'e', 'f', 'h', False, 1, 2, None]
    ):

        assert flat_iterator_item == check_item

    assert list(flat_generator(list_of_lists_1)) == ['a', 'b', 'c', 'd', 'e', 'f', 'h', False, 1, 2, None]

    assert isinstance(flat_generator(list_of_lists_1), types.GeneratorType)


if __name__ == '__main__':
    test_2()
```
